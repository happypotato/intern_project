```python
results_small_std = pd.read_csv('results/pos_mean_sd0.1.csv',header=None,sep=';')
results_large_std = pd.read_csv('results/pos_mean_sd1.0.csv',header=None,sep=';')
results_neg_mean_small_std = pd.read_csv('results/neg_mean_sd0.1.csv',header=None,sep=';')
results_neg_mean_large_std = pd.read_csv('results/neg_mean_sd1.0.csv',header=None,sep=';')
results_neg_mean_med_std = pd.read_csv('results/neg_mean_sd0.5.csv',header=None,sep=';')
```

```python

def plot_results(results0,title):
    results = results0.copy(True)
    results.columns = ['model','ctime','vals']
    results['vals_plot'] = results.vals.map(lambda x: [float(i) for i in x.split(',')])

    bayes = results[results.model == 'bayes']
    flr = results[results.model == 'flr']
    expert = results[results.model == 'expert']

    bayes_sims = dict(zip(range(len(bayes)),bayes.vals_plot))
    flr_sims = dict(zip(range(len(bayes)),flr.vals_plot))
    expert_sims = dict(zip(range(len(bayes)),expert.vals_plot))

    fig, ax = plt.subplots()
    
    #plot([0,len(bayes)], [.5,.5], 'k--',color='red',label='Bayesian LR')
    for s in bayes_sims:
        bayes_label = None
        flr_label = None
        expert_label = None
        if s == 0:
            bayes_label = 'Bayesian LR'
            flr_label = 'Frequentist LR'
            expert_label = 'Expert System'
        plot(bayes_sims[s],color='red',alpha=0.3,label=bayes_label)
        plot(flr_sims[s],color='blue',alpha = 0.3,label=flr_label)
        plot(expert_sims[s],color='green',alpha = 0.2,label=expert_label)  
        
    plt.legend(loc="lower right",fontsize=14)


    ax.set_ylabel("Eventual AUC",fontsize=14)
    ax.set_xlabel("Month",fontsize=14)
    
    plt.title(title,fontsize=16)
    
    plt.gcf().set_size_inches(10, 10);
        
```

```python
 plot_results(results_small_std,'Bayesian Prior Mean = 1, Std = 0.1') 
```

```python
 plot_results(results_large_std,'Bayesian Prior Mean = 1, Std = 1') 
```

```python
 plot_results(results_neg_mean_small_std,'Bayesian Prior Mean = -1, Std = 0.1') 
```

```python
 plot_results(results_neg_mean_large_std,'Bayesian Prior Mean = -1, Std = 1.0') 
```

```python
 plot_results(results_neg_mean_med_std,'Bayesian Prior Mean = -1, Std = 0.5') 
```
